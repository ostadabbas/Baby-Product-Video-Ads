# merge all features

```python
stats = content_stats.merge(color_stats, on='ad_id', how='left')
```

```python
len(stats)
```

```python
stats.head(2)
```

```python
def split_ad_id(ad_id):
    if '+' in ad_id[0:5]:
        return ad_id.split('+')[0]
    elif '-' in ad_id[0:5]:
        return ad_id.split('-')[0]
    else:
        return ad_id  # if neither '+' nor '-' is present, return the original ad_id
```

```python
stats['VideoNumber'] = stats['ad_id'].apply(split_ad_id)
```

```python
stats.head(2)
```

```python
stats['VideoNumber'] = stats['VideoNumber'].astype(float)
```
